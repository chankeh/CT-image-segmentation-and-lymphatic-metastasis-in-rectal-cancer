```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier as RF
import tensorflow
import keras
from sklearn.decomposition import PCA
```

```python
f_1 = np.load('/Users/liuxiaoyu/Downloads/npys/feature_one.npy')
f_2 = np.load('/Users/liuxiaoyu/Downloads/npys/feature_two.npy')
f_3 = np.load('/Users/liuxiaoyu/Downloads/npys/feature_three.npy')
feature = np.hstack((f_1,f_2,f_3))
```

```python
path = '/Users/liuxiaoyu/Desktop/clinic.csv'
clinic = pd.read_csv(path)
clin = clinic.as_matrix()
clin.shape
sex = clin[ : , 0:1]
age = clin[ : , 1:2]
sex_number = np.zeros([107,2])
for i in range(107):
    if sex[i] == 'm':
        sex_number[i,0] = 1
    else:
        sex_number[i,1] = 1


x = np.hstack((feature, age, sex_number))

F_rf = np.zeros([64,20])
frac = np.zeros([64, 20])
for j in range(1,64):
    for k in range(81,100):
        pca = PCA(n_components = j)
        x_new = pca.fit_transform(x)
        x_train = x_new[0:k,:]
        x_test = x_new[k:107,:]
        y_train = clin[0:k, 2:3]
        y_test = clin[k:107, 2:3]
        clf = RF(max_depth = 2, random_state = 0)
        clf.fit(x_train, y_train)
        y_pred = clf.predict(x_test)
        count = 0
        pn = y_test.shape[0]
        for i in range(pn):
            if y_test[i] == y_pred[i]:
                count = count +1
        frac[j, k-80] = count/pn
        
        TN, TP, FP, FN = 0,0,0,0 
        for l in range(pn):
            if (y_test[l] == '+') & (y_pred[l] == '+'):
                TP = TP +1
            if (y_test[l] == '+') & (y_pred[l] == '-'):
                FN = FN +1
            if (y_test[l] == '-') & (y_pred[l] == '+'):
                FP = FP + 1
            else:
                TN = TN + 1
        if (TP+FP != 0) & (TP+FN != 0):
            P = TP/(TP+FP)
            R = TP/(TP+FN)
            if P + R != 0:
                F_rf[j,k-80] = 2*P*R/(P+R)
        else:
            F_rf[j,k-80]  = 0 
```

```python
a = np.where(F_rf>0.7)[1]
b = np.where(F_rf>0.7)[0]

print(a+80,b)
w = np.where(a==16)
w [0]
Fscore_RF = F_rf[b[w[0]], 16]
print(Fscore_RF)
```
